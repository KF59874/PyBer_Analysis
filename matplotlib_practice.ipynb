```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot (x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()

# This version should only be used once per session because it could 
# otherwise distort following graphs that also use plt.show() function
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston', linewidth=2.0, color="green" , marker="D")
# Other markers: https://matplotlib.org/stable/api/markers_api.html

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

#To show info only from Jan to June use this
# plt.xlim("Jan","June")
# or this
# plt.xlim(0,5)

# Set the y limit between 0 and 45.
plt.ylim(0, 45)

# Create a title.
plt.title("PyBer Fare by Month")

#Add a grid
plt.grid()

#plt.figure(figsize=(20, 10))


# Add the legend.
plt.legend()

# Other possibilites for markers: *, >, ^, +
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.scatter(y_axis, x_axis, c="red")
plt.legend("Chicago")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.title("PyBer Fare by Month")
```

```python
# Create a bubble chart 

plt.scatter(x_axis, y_axis, s=y_axis)

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```
