# Pyber Challenge

### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import datetime 
import os 

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# 2. Get the total drivers for each city type
total_drivers = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers
```

```python
#  3. Get the total amount of fares for each city type
total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride = total_fares /total_rides
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
average_fare_per_driver = total_fares/total_drivers
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 
pbyer_summary_df = pd.DataFrame()

pbyer_summary_df["Total Rides"] = total_rides
pbyer_summary_df["Total Drivers"] = total_drivers
pbyer_summary_df["Total Fares"] = total_fares
pbyer_summary_df["Average Fare per Ride"] = average_fare_per_ride
pbyer_summary_df["Average Fare per Driver"] = average_fare_per_driver

pbyer_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pbyer_summary_df.index.name = None
```

```python
#  8. Format the columns.
pbyer_summary_df["Total Rides"] = pbyer_summary_df["Total Rides"].map("{:,}".format)
pbyer_summary_df["Total Drivers"] = pbyer_summary_df["Total Drivers"].map("{:,}".format)
pbyer_summary_df["Total Fares"] = pbyer_summary_df["Total Fares"].map("${:,.2f}".format)
pbyer_summary_df["Average Fare per Ride"] = pbyer_summary_df["Average Fare per Ride"].map("${:.2f}".format)
pbyer_summary_df["Average Fare per Driver"] = pbyer_summary_df["Average Fare per Driver"].map("${:.2f}".format)

pbyer_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
sum_of_fares = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
sum_of_fares
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
sum_of_fares_df = pd.DataFrame(sum_of_fares)
sum_of_fares_df
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
sum_of_fares_df = sum_of_fares_df.reset_index()
sum_of_fares_df.head(10)
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
sum_of_fares_pivot = sum_of_fares_df.pivot(index="date", columns="type", values="fare")
sum_of_fares_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
sum_of_fares_pivot_df = sum_of_fares_pivot.loc['2019-01-01':'2019-04-28']
sum_of_fares_pivot_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
sum_of_fares_pivot_df.index = pd.to_datetime(sum_of_fares_pivot_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
sum_of_fares_pivot_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
sum_of_fares_pivot_week = sum_of_fares_pivot_df.resample("W").sum()
sum_of_fares_pivot_week
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
sum_of_fares_pivot_week.plot(figsize=(14, 4))

# Add graph properties
plt.title("Total Fare by City Type", fontsize=18)
plt.ylabel("Fare ($USD)", fontsize=15)
plt.xlabel("Month", fontsize=15)

# Create a legend
lgnd = plt.legend(fontsize="10", title="City Type")
lgnd.get_title().set_fontsize(10)

# Save figure
plt.savefig("analysis/PyBer_fare_summary.png")
```
